```python
%matplotlib inline
import matplotlib.pyplot as plt
from mxnet import image

img = image.imdecode(open('./img/cat1.jpg','rb').read())
plt.imshow(img.asnumpy())
```

定义一个辅助函数，给定输入图片`img`的增强方法`aug`，运行多次给出结果

```python
def apply(img, aug, n=3):
    _, figs = plt.subplots(n, n, figsize=(8,8))
    for i in range(n):
        for j in range(n):
            ''' 转成float, 一是因为aug需要float类型数据来变化。
            二是这里会有一次copy操作，因为有些aug直接通过改写输入
            (而不是新建输出)获取性能的提升 '''
            x = img.astype('float32')
            # clip保证在要求范围之内
            y = aug(x).clip(0, 254)
            figs[i][j].imshow(y.asnumpy()/255.0)
            # 显示浮点图片时imshow要求在[0,1]之间
            figs[i][j].axes.get_xaxis().set_visible(False)
            figs[i][j].axes.get_yaxis().set_visible(False)
```

### 变形

水平方向翻转图片是最早也是最广泛的一种增强

```python
# 以.5的概率翻转
aug = image.HorizontalFlipAug(.5)
apply(img, aug)
```

样例图片里我们关心的猫在图片正中间，但因为物体不一定在图片正中央。前面提到过池化层能弱化卷积对目标敏感度，但不能完全解决这个问题。常用的增强方法是随机截取其中一小块。

注意到随机剪裁图片会缩小输入的形状。如果原始图片过小，导致没有空间进行随机剪裁，通常做法是把图片放大到足够大。如果原始图片足够大，建议不要事先将它们裁剪到网络所需大小。

```python
# 随机裁剪一块 200 x 200 的区域
aug = image.RandomCropAug([200,200])
apply(img, aug)
```

我们也可以随机剪裁一块随机大小

```python
# 随机剪裁 要求保留至少 0.1的区域，随机长宽比在 .5和 2之间
# 最后将结果resize到 200x200
aug = image.RandomSizedCropAug((200,200), .1, (.5,2))
apply(img, aug)
```

### 颜色变化

形状变化外的另一大类是变化颜色

```python
# 随机将亮度增加或者减小在0-50%间的一个量
aug = image.BrightnessJitterAug(.5)
apply(img, aug)
```

```python
# 随机色调变化
aug = image.HueJitterAug(.5)
apply(img, aug)
```

或者用使用`RandomColorJitter`来一起使用。

```python
from mxnet.gluon.data.vision import transforms
color_aug = transforms.RandomColorJitter(
    brightness=.5, contrast=.5, saturation=.5, hue=.5)
apply(img, color_aug)
```

### 数据读取

定义一个辅助函数对图片按顺序应用数个增强

```python
def apply_aug_list(img, augs):
    for f in augs:
        img = f(img)
    return img
```
